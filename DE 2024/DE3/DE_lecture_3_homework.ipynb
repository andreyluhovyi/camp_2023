# Task 1

Реалізуй функцію `numbers_handler`, яка приймає список чисел numbers і три колбеки:

    * before - колбек, який першим викликається для кожного з чисел
    * action - викликається другим для кожного з чисел
    * after - викликається останнім для кожного з чисел

Кожен із колбеків має єдиний параметр - число `number`.

Функція повинна викликати всі три колбеки у вказаному порядку для кожного з чисел.

Крім того, реалізуй значення за замовчуванням для цих колбеків.

Для `before` напиши функцію `new_number`, яка приймає число та виводить рядок `Received a new number: {number}` за допомогою функції `print`.

Для `action` напиши функцію `is_positive`, яка приймає число та виводить рядок `{number} is a positive number` якщо число `number` - додатне, рядок `Zero`, якщо число дорівнює нулю, і рядок `{number} is a negative number`, якщо число від'ємне.

Для `after` напиши функцію `print_bye`, яка просто виводить рядок `Bye!`.

**Example:**


```python
numbers_handler([1, -1, 0]) # викликаємо функцію зі значеннями функцій за замовчуванням
# на екран виводиться:
# Received a new number: 1
# 1 is a positive number
# Bye!
# Received a new number: -1
# -1 is a negative number
# Bye!
# Received a new number: 0
# Zero
# Bye!


def print_hello(number):
    print("Hello!")

def some_action(number):
    print("Action!")

numbers_handler([10], action=some_action, before=print_hello)
# Викликаємо функцію зі своїми функціями
# на екран виводиться:
# Hello!
# Action!
# Bye!
```



````python
def numbers_handler(numbers, before=new_number, action=is_positive, after=print_bye):
    """
    Function processes a list of numbers by calling three callbacks for each number: before, action, and after.

    Args:
        numbers: list of numbers
        before: callback function called first
        action: callback function called second
        after: callback function called last

    Returns:
        None
    """
    
    for number in numbers:
        before(number)
        action(number)
        after(number)


def new_number(number):
    print(f'Received a new number: {number}')

def is_positive(number):
    if number > 0:
        print(f'{number} is a positive number')
    elif number == 0:
        print('Zero')
    else:
        print(f'{number} is a negative number')

def print_bye(number):
    print('Bye!')

# Test using example above
numbers_handler([1, -1, 0])
````

Наприкінці навчального року час виставляти річні. За рік кожен з учнів отримує кілька оцінок. Річний бал дорівнює середньому арифметичному, округленому до найближчого цілого.

Напиши функцію `get_average_mark`, яка приймає ім'я студента та довільну ненульову кількість його оцінок. Вона має повернути рядок-результат, у якому міститься підсумковий бал у форматі `{student_name} got {student_mark}`.

**Example**:

```python
get_average_mark("Danylo", 11) == "Danylo got 11"
get_average_mark("Oleksii", 12, 10, 11, 11, 11) == "Oleksii got 11"
get_average_mark("Ivan", 10, 11, 12, 9, 12, 12) == "Ivan got 11"
```

````python
def get_average_mark(student_name, rating):
  """
  Calculates and returns a formatted string representing the average mark for a student.

  Args:
      student_name: The name of the student (string).
      rating: A list of integers containing the student's marks.

  Returns:
      A string formatted as "{student_name} got {average_mark}"
      where {average_mark} is the average mark rounded to the nearest whole number.
  """

  student_mark = f"{sum(rating) / len(rating):.0f}"  # Calculate and format average

  return f"{student_name} got {student_mark}"  # Combine name and average mark

# Test using example above
print(get_average_mark("Danylo", [11]))
print(get_average_mark("Oleksii", [12, 10, 11, 11, 11]))
print(get_average_mark("Ivan", [10, 11, 12, 9, 12, 12]))
````

# Task 3

Напиши функцію `message_people`, яка приймає список імен `people`, створює і повертає функцію `print_message`.

Внутрішня функція приймає строку `message` і в залежності від того, чому дорівнює `message` `('hello'/'meeting'/'bye')` виводить результати, які показані в прикладі, функція `print_message` не повинна нічого повертати.

```python
mes_people = message_people(["Alex", "Robert", "Tom"])
mes_people('hello')
# Hello, Alex, Robert, Tom, nice to see you!
mes_people('meeting')
# Alex, Robert, Tom, we have a meeting in an hour, don't be late!
mes_people('bye')
# Bye Alex, Robert, Tom, see you tomorrow!
```

````python
def message_people(people):
  """
  This function creates and returns another function named 'print_message'
  that takes a message argument and prints greetings based on the message.

  Args:
      people: A list containing names of people (strings).

  Returns:
      A function named 'print_message'.
  """

  messages = {
      'hello': f"Hello, {', '.join(people)}, nice to see you!",
      'meeting': f"{', '.join(people)}, we have a meeting in an hour, don't be late!",
      'bye': f"Bye {', '.join(people)}, see you tomorrow!"
  }

  def print_message(message):
    """
    This function prints a greeting message for the people in the 'people' list.
    The message to be printed depends on the 'message' argument.

    Args:
        message: A string representing the type of message ('hello', 'meeting', or 'bye').
    """
    print(messages.get(message, "Invalid message"))

  return print_message

# Test using example above
mes_people = message_people(["Alex", "Robert", "Tom"])
mes_people('hello')
mes_people('meeting')
mes_people('bye')
````

Числа Фібоначчі - це числова послідовність, у якій перші два елементи дорівнюють нулю та одиниці, а кожен наступний дорівнює сумі двох попередніх.

    F₀ = 0,
    F₁ = 1,
    Fᵢ = Fᵢ₋₁ + Fᵢ₋₂

Ось перші елементи цієї послідовності: `0, 1, 1, 2, 3, 5, 8, 13...`

Детальніше числа Фібоначчі можна прочитати за цим посиланням. https://www.techtarget.com/whatis/definition/Fibonacci-sequence

Напишіть генератор `fibonacci_generator`, що приймає число n і рахує перші n чисел Фібоначчі.

**Example:**

```python

fib_generator = fibonacci_generator(6)
for fib_number in fib_generator:
    print(fib_number)
# 0
# 1
# 1 (1 + 0)
# 2 (1 + 1)
# 3 (2 + 1)
# 5 (3 + 2)

```



````python
def fibonacci_generator(n, a=0, b=1):
  """
  Function-generator that yields the first n Fibonacci numbers.

  This function utilizes recursion to efficiently generate Fibonacci numbers.

  Args:
    n: The number of Fibonacci numbers to generate.
    a: The first Fibonacci number (default: 0).
    b: The second Fibonacci number (default: 1).

  Yields:
    The next Fibonacci number.
  """
  if n > 0:
    yield a
    yield from fibonacci_generator(n-1, b, a+b)  # Recursive call

# Test using example above
fib_generator = fibonacci_generator(6)
for fib_number in fib_generator:
  print(fib_number)
````

## class Person


You have a list of dicts `people`, every dict means
a **person**, it has keys: `name`, `age`, 
`wife`/`husband` - depends on person is male or

Create class `Person`. It's constructor takes

Write function `create_person_list`, this function

`Person` instances instead of dicts.

**Note:**

If **person's** key `wife`/`husband` is not 
`None` - `create_person_list` should add

to its instance. This attribute should

````python
class Person:
  """
  This class represents a person with a name, age, and optional spouse information.
  """

  people = {}

  def __init__(self, name, age):
    """
    Initializes a new Person object with the given name and age.

    Args:
      name: The name of the person.
      age: The age of the person.
    """
    self.name = name
    self.age = age
    self.wife = None
    self.husband = None
    type(self).people[name] = self 

  @classmethod
  def get_person_by_name(cls, name):
    """
    Retrieves a Person object from the central registry by name.

    Args:
      name: The name of the person to retrieve.

    Returns:
      The Person object with the given name, or None if not found.
    """
    return cls.people.get(name)

def create_person_list(people):
  """
  Creates a list of Person objects from a list of dictionaries representing people.

  Args:
    people: A list of dictionaries, where each dictionary represents a person
      with keys like 'name', 'age', optionally 'wife', and 'husband'.

  Returns:
    A list of Person objects created based on the input data.
  """
  person_list = []
  for p in people:
    person_instance = Person(p['name'], p['age'])
    wife_name = p.get('wife')
    if wife_name:
      wife = Person.get_person_by_name(wife_name)
      person_instance.wife = wife
      if wife:
        wife.husband = person_instance
    husband_name = p.get('husband')
    if husband_name:
      husband = Person.get_person_by_name(husband_name)
      person_instance.husband = husband
      if husband:
        husband.wife = person_instance
    person_list.append(person_instance)
  return person_list
````

````python
# Test using example above

people = [
  {'name': 'Ross', 'age': 30, 'wife': 'Rachel'},
  {'name': 'Joey', 'age': 29, 'wife': None},
  {'name': 'Rachel', 'age': 28, 'husband': 'Ross'}
]

person_list = create_person_list(people)

print(isinstance(person_list[0], Person))  # True
print(person_list[0].name == 'Ross')
print(person_list[0].wife is person_list[2])  # True
print(person_list[0].wife.name == 'Rachel')

print(person_list[1].name == 'Joey')
print(person_list[1].wife)  # None

print(isinstance(person_list[2], Person))  # True
print(person_list[2].name == 'Rachel')
print(person_list[2].husband is person_list[0])  # True
print(person_list[2].husband.name == 'Ross')
print(person_list[2].husband.wife is person_list[2])  # True

print(Person.people)
````
