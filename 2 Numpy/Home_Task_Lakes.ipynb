## Task about lakes

- Generate 20 random variable from 1 to 100
- Draw the plot of the  sequence 
- Lets assume each point represent the height and so all plotting is 2d mountains. 
- Then consider the unlimited rain from above - cavities become lakes full of water. 
- Determine the deepest lake

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def find_lake_boundary(boundary, direction, terrain):
    while boundary > 0 and boundary < len(terrain) - 1 and terrain[boundary + direction] > terrain[boundary]:
        boundary += direction
    return boundary

def find_deepest_lake(num_attempts=3):
    for _ in range(num_attempts):
        terrain = np.random.randint(1, 100, 20)

        lake_depths = {}
        for point in range(1, len(terrain) - 1):
            if terrain[point - 1] > terrain[point] <= terrain[point + 1]:
                left_boundary = find_lake_boundary(point - 1, -1, terrain)
                right_boundary = find_lake_boundary(point + 1, 1, terrain)
                depth = min(terrain[left_boundary], terrain[right_boundary]) - terrain[point]
                lake_depths[depth] = point

        deepest_depth = max(lake_depths)
        deepest_lake = lake_depths[deepest_depth]

        left_boundary = find_lake_boundary(deepest_lake, -1, terrain)
        right_boundary = find_lake_boundary(deepest_lake, 1, terrain)

        plt.figure()
        plt.plot(terrain)
        plt.plot(range(left_boundary, right_boundary + 1), terrain[left_boundary:right_boundary + 1], color='red')
        plt.xlabel('Index')
        plt.ylabel('Value')
        plt.title(f'Deepest Lake - Trial {_ + 1}')
        plt.show()

find_deepest_lake(num_attempts=3)
```
