###  Data

```python
import os
os.environ["OMP_NUM_THREADS"] = '2'
```

```python
from sklearn.datasets import make_blobs
import numpy as np
import matplotlib.pyplot as plt
```

```python
X, y = make_blobs(n_samples = 500, n_features = 2, centers = 5, 
                           cluster_std = 0.6, random_state = 0)
colors = np.array([plt.cm.Spectral(val)
          for val in np.linspace(0, 1, len(set(y)))])
plt.figure(figsize=(8,6))

plt.scatter(X[:,0], X[:,1], c= colors[y], s= 20) 
```

```python
colors = np.array(['purple', 'teal', 'magenta', 'indigo', 'coral', 'lime'])
colors = np.concatenate([colors, np.repeat('black', 100)])
```


<font color = green >

###  Agglomerative clustering using `complete` linkage

</font>

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
clf = AgglomerativeClustering(n_clusters=5, linkage= 'complete')
predicted = clf.fit_predict(X)
```

```python
plt.figure(figsize=(9,9))
plt.scatter(
    X[:,0],
    X[:,1],
    c= colors[predicted],
    marker= '.',
    s=100
) 
plt.title ('Agglomerative Clustering with 5 clusters(Complete Linkage)')
plt.xlabel('X-axis label')
plt.ylabel('Y-axis label')
```


<font color = green >

###  Agglomerative clustering using `single` linkage

</font>

```python
clf = AgglomerativeClustering(n_clusters= 4,linkage="single")
predicted = clf.fit_predict(X)
```

```python
plt.figure(figsize=(9,9))
plt.scatter(
    X[:,0],
    X[:,1],
    c= colors[predicted],
    marker= '.',
    s=100
) 
plt.title ('Agglomerative Single Clustering with 5 clusters')
plt.xlabel('X-axis label')
plt.ylabel('Y-axis label')
```


<font color = green >

###  DBSCAN

</font>

```python
from sklearn.cluster import DBSCAN
```

```python
dbscan = DBSCAN(eps = 0.4, min_samples = 5)
predicted = dbscan.fit_predict(X)
```

```python
plt.figure(figsize=(9,9))
plt.scatter(
    X[:,0],
    X[:, 1], 
    c= colors[predicted], 
    marker= '.', 
    s=100
) 
plt.title ('DBSCAN Clustering Results')
plt.xlabel('X-axis label')
plt.ylabel('Y-axis label')
```

<font color = green >

###  the best k for Kmeans

</font>

```python
from sklearn.cluster import KMeans
```

```python
k_range = range(1, 10)
scores = [KMeans(n_clusters=k).fit(X).inertia_ for k in k_range]
```

```python
plt.figure(figsize=(9, 9))
plt.plot(k_range,
         scores,
         marker='o',
         color='b'
)
```
